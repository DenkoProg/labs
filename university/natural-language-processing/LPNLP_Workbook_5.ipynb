```python
import json
import numpy as np
from pathlib import Path

def load_array(path):
    with open(path, "r", encoding="utf-8") as f:
        return np.array(json.load(f), dtype=np.float32)

def load_vocab(path):
    with open(path, "r", encoding="utf-8") as f:
        vocab = json.load(f)
    if isinstance(vocab, list):
        id2tok = vocab
        tok2id = {t: i for i, t in enumerate(vocab)}
    else:
        tok2id = vocab
        id2tok = [None] * (max(tok2id.values()) + 1)
        for t, i in tok2id.items():
            id2tok[i] = t
    return tok2id, id2tok

class SimpleRNN:
    def __init__(self, U_h, W_h, b_h, W_y, E, tok2id, id2tok):
        self.U_h, self.W_h, self.b_h, self.W_y, self.E = U_h, W_h, b_h, W_y, E
        self.tok2id, self.id2tok = tok2id, id2tok
        self.id_start = tok2id["["]
        self.id_end = tok2id["]"]
        self.H = U_h.shape[0]

    def step(self, prev_id, h):
        e = self.E[prev_id]
        h = np.tanh(self.U_h @ h + self.W_h @ e + self.b_h)
        logits = self.W_y @ h
        return int(np.argmax(logits)), h

    def generate(self, max_len=64):
        h = np.zeros(self.H, np.float32)
        prev, ids = self.id_start, []
        for _ in range(max_len):
            nxt, h = self.step(prev, h)
            if nxt == self.id_end: break
            ids.append(nxt)
            prev = nxt
        toks = [self.id2tok[i] for i in ids]
        return "".join(toks), ids

U_h = load_array("/content/U_h.weight.json")
W_h = load_array("/content/W_h.weight.json")
b_h = load_array("/content/W_h.bias.json")
W_y = load_array("/content/W_y.weight.json")
E   = load_array("/content/embedding.weight.json")
tok2id, id2tok = load_vocab(Path("/content/vocab.json"))

model = SimpleRNN(U_h, W_h, b_h, W_y, E, tok2id, id2tok)
text, ids = model.generate(64)

print("len:", len(ids))
print("tokens:", [id2tok[i] for i in ids[:50]])
print("TEXT:", text)
```
